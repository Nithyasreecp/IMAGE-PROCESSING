```python
# Read image
img = cv2.imread("/content/WhatsApp Image 2025-04-02 at 14.04.41_a0df696f.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR â†’ RGB
```

```python
#Addition (make image brighter)
brighter = cv2.add(img, 50)   # Add 50 to all pixels
plt.imshow(brighter)
plt.title("Brighter Image")
plt.axis("off")
plt.show()
```

```python
#Subtraction (make image darker)
darker = cv2.subtract(img, 50)   # Subtract 50 from all pixels
plt.imshow(darker)
plt.title("Darker Image")
plt.axis("off")
plt.show()
```

```python
#Multiplication (increase contrast)
contrast = cv2.multiply(img, 1.5)   # Multiply pixels by 1.5
plt.imshow(contrast)
plt.title("High Contrast")
plt.axis("off")
plt.show()
```

```python
#Division (reduce intensity)
reduced = cv2.divide(img, 2)   # Divide all pixels by 2
plt.imshow(reduced)
plt.title("Reduced Intensity")
plt.axis("off")
plt.show()
```

```python
#aithamatic operations in mannual way using for loop
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Step 1: Load image using PIL
img = Image.open("/content/WhatsApp Image 2025-04-02 at 14.04.41_a0df696f.jpg")  # Replace with your image path
img = np.array(img)  # Convert to NumPy array
```

```python
# Step 2: Create an empty array for result
manual_add = np.zeros_like(img)
```

```python
# Step 3: Manually add 50 to each pixel
rows, cols, channels = img.shape
for i in range(rows):
    for j in range(cols):
        for c in range(channels):
            val = img[i, j, c] + 50
            # Clip manually to 255
            if val > 255:
                val = 255
            manual_add[i, j, c] = val
```

```python

# Step 4: Convert back to image and display
result_img = Image.fromarray(manual_add.astype(np.uint8))

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(result_img)
plt.title("Brightened Image (+50 manually)")
plt.axis("off")
plt.show()
```
